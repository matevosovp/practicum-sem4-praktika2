```python
def overview(df: pd.DataFrame, name: str) -> None:
    print(f"\n===== {name} =====")
    print("shape:", df.shape)
    display(df.head())
    print("\nТипы:")
    display(df.dtypes.to_frame("dtype"))
    print("\nПропуски:")
    display(df.isna().sum().to_frame("missing_cnt"))
    print("\nДоля пропусков:")
    display((df.isna().mean().mul(100)).round(2).to_frame("missing_pct"))

overview(category_tree, "category_tree")
overview(events, "events")
overview(item_properties, "item_properties")
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_colwidth", 200)

DATA_DIR = Path(".")

CATEGORY_PATH = DATA_DIR / "category_tree.csv"
EVENTS_PATH = DATA_DIR / "events.csv"
ITEM_PROP_1_PATH = DATA_DIR / "item_properties_part1.csv"
ITEM_PROP_2_PATH = DATA_DIR / "item_properties_part2.csv"
OUTPUT_PATH = DATA_DIR / "EDA_FINDINGS.md"


def safe_pct(num, den):
    if den == 0:
        return 0.0
    return round(100 * num / den, 2)


def series_quantile_safe(s: pd.Series, q: float):
    if len(s) == 0:
        return np.nan
    return s.quantile(q)


def top_n_share(counts: pd.Series, n: int = 10):
    if counts.empty:
        return 0.0
    total = counts.sum()
    if total == 0:
        return 0.0
    return round(100 * counts.sort_values(ascending=False).head(n).sum() / total, 2)


def format_dt(x):
    if pd.isna(x):
        return "NA"
    return pd.Timestamp(x).strftime("%Y-%m-%d %H:%M:%S")


def format_num(x):
    if pd.isna(x):
        return "NA"
    if isinstance(x, (float, np.floating)):
        if float(x).is_integer():
            return f"{int(x):,}".replace(",", " ")
        return f"{x:,.2f}".replace(",", " ")
    if isinstance(x, (int, np.integer)):
        return f"{int(x):,}".replace(",", " ")
    return str(x)


def detect_parent_child_columns(df: pd.DataFrame):
    cols = [c.lower() for c in df.columns]
    parent_col = None
    child_col = None

    for c in df.columns:
        cl = c.lower()
        if "parent" in cl:
            parent_col = c
        if "child" in cl:
            child_col = c

    if parent_col is None and len(df.columns) >= 1:
        parent_col = df.columns[0]
    if child_col is None and len(df.columns) >= 2:
        child_col = df.columns[1]

    return parent_col, child_col


category_tree = pd.read_csv(CATEGORY_PATH)
events = pd.read_csv(EVENTS_PATH)
item_properties_1 = pd.read_csv(ITEM_PROP_1_PATH)
item_properties_2 = pd.read_csv(ITEM_PROP_2_PATH)
item_properties = pd.concat([item_properties_1, item_properties_2], ignore_index=True)

events["timestamp"] = pd.to_datetime(events["timestamp"], unit="ms", errors="coerce")
item_properties["timestamp"] = pd.to_datetime(item_properties["timestamp"], unit="ms", errors="coerce")

parent_col, child_col = detect_parent_child_columns(category_tree)

category_rows = len(category_tree)
category_missing = category_tree.isna().sum().sum()
category_unique_parents = category_tree[parent_col].nunique(dropna=True) if parent_col in category_tree.columns else np.nan
category_unique_children = category_tree[child_col].nunique(dropna=True) if child_col in category_tree.columns else np.nan

events_rows, events_cols = events.shape
events_missing = events.isna().sum().sum()
unique_users = events["visitorid"].nunique(dropna=True)
unique_items_events = events["itemid"].nunique(dropna=True)
unique_transactions = events["transactionid"].nunique(dropna=True)

item_props_rows, item_props_cols = item_properties.shape
item_props_missing = item_properties.isna().sum().sum()
unique_items_props = item_properties["itemid"].nunique(dropna=True)
unique_properties = item_properties["property"].nunique(dropna=True)
unique_property_values = item_properties["value"].nunique(dropna=True)

events_min_ts = events["timestamp"].min()
events_max_ts = events["timestamp"].max()
props_min_ts = item_properties["timestamp"].min()
props_max_ts = item_properties["timestamp"].max()

event_counts = events["event"].value_counts(dropna=False)
event_shares = (event_counts / len(events) * 100).round(2)

views = int(event_counts.get("view", 0))
carts = int(event_counts.get("addtocart", 0))
transactions = int(event_counts.get("transaction", 0))

user_activity = events.groupby("visitorid").size().rename("events_per_user")
item_activity = events.groupby("itemid").size().rename("events_per_item")

median_events_per_user = user_activity.median()
p95_events_per_user = series_quantile_safe(user_activity, 0.95)
p99_events_per_user = series_quantile_safe(user_activity, 0.99)

median_events_per_item = item_activity.median()
p95_events_per_item = series_quantile_safe(item_activity, 0.95)
p99_events_per_item = series_quantile_safe(item_activity, 0.99)

top_10_items_share = top_n_share(item_activity, 10)
top_100_items_share = top_n_share(item_activity, 100)

user_item_pairs = (
    events.groupby(["visitorid", "itemid", "event"])
    .size()
    .rename("cnt")
    .reset_index()
)

pair_event_pivot = (
    user_item_pairs.pivot_table(
        index=["visitorid", "itemid"],
        columns="event",
        values="cnt",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
)

for col in ["view", "addtocart", "transaction"]:
    if col not in pair_event_pivot.columns:
        pair_event_pivot[col] = 0

pair_event_pivot["has_view"] = (pair_event_pivot["view"] > 0).astype(int)
pair_event_pivot["has_addtocart"] = (pair_event_pivot["addtocart"] > 0).astype(int)
pair_event_pivot["has_transaction"] = (pair_event_pivot["transaction"] > 0).astype(int)

pairs_total = len(pair_event_pivot)
pairs_with_view = int(pair_event_pivot["has_view"].sum())
pairs_with_cart = int(pair_event_pivot["has_addtocart"].sum())
pairs_with_transaction = int(pair_event_pivot["has_transaction"].sum())

cart_after_view_rate = round(
    pair_event_pivot.loc[pair_event_pivot["has_view"] == 1, "has_addtocart"].mean() * 100, 2
) if pairs_with_view > 0 else 0.0

transaction_after_view_rate = round(
    pair_event_pivot.loc[pair_event_pivot["has_view"] == 1, "has_transaction"].mean() * 100, 2
) if pairs_with_view > 0 else 0.0

transaction_after_cart_rate = round(
    pair_event_pivot.loc[pair_event_pivot["has_addtocart"] == 1, "has_transaction"].mean() * 100, 2
) if pairs_with_cart > 0 else 0.0

repeat_views_share = round((pair_event_pivot["view"] > 1).mean() * 100, 2) if len(pair_event_pivot) > 0 else 0.0
repeat_carts_share = round((pair_event_pivot["addtocart"] > 1).mean() * 100, 2) if len(pair_event_pivot) > 0 else 0.0

property_stats = (
    item_properties.groupby("property")
    .agg(
        rows=("itemid", "size"),
        unique_items=("itemid", "nunique"),
        unique_values=("value", "nunique")
    )
    .sort_values("rows", ascending=False)
    .reset_index()
)

top_properties_md = ""
if not property_stats.empty:
    top_properties = property_stats.head(15).copy()
    top_properties_md = top_properties.to_markdown(index=False)
else:
    top_properties_md = "No property stats available"

item_property_versions = (
    item_properties.groupby(["itemid", "property"])
    .size()
    .rename("versions_cnt")
    .reset_index()
)

median_versions_per_item_property = item_property_versions["versions_cnt"].median() if not item_property_versions.empty else np.nan
p95_versions_per_item_property = series_quantile_safe(item_property_versions["versions_cnt"], 0.95) if not item_property_versions.empty else np.nan

items_only_in_events = len(set(events["itemid"].dropna().unique()) - set(item_properties["itemid"].dropna().unique()))
items_only_in_props = len(set(item_properties["itemid"].dropna().unique()) - set(events["itemid"].dropna().unique()))

daily_events = (
    events.assign(date=events["timestamp"].dt.date)
    .groupby(["date", "event"])
    .size()
    .rename("cnt")
    .reset_index()
)

daily_events_md = "No daily events summary"
if not daily_events.empty:
    daily_pivot = daily_events.pivot(index="date", columns="event", values="cnt").fillna(0)
    daily_summary = pd.DataFrame({
        "metric": ["days", "avg_daily_views", "avg_daily_addtocart", "avg_daily_transaction"],
        "value": [
            len(daily_pivot),
            round(daily_pivot["view"].mean(), 2) if "view" in daily_pivot.columns else 0,
            round(daily_pivot["addtocart"].mean(), 2) if "addtocart" in daily_pivot.columns else 0,
            round(daily_pivot["transaction"].mean(), 2) if "transaction" in daily_pivot.columns else 0,
        ]
    })
    daily_events_md = daily_summary.to_markdown(index=False)

implications = []

if carts > 0 and transactions > 0:
    implications.append(
        f'В качестве основного позитивного сигнала разумно использовать "addtocart", '
        f'а покупки применять как более сильный дополнительный сигнал и для пост-фильтрации уже купленных товаров'
    )
else:
    implications.append(
        'Нужно отдельно проверить качество разметки событий, так как позитивные сигналы могут быть слишком редкими'
    )

if top_100_items_share >= 20:
    implications.append(
        f"Наблюдается заметная концентрация популярности товаров: топ-100 товаров дают {top_100_items_share}% всех событий. "
        f"Популярностный baseline обязателен"
    )
else:
    implications.append(
        "Концентрация популярности умеренная, поэтому одного popularity baseline будет недостаточно"
    )

if median_versions_per_item_property and not pd.isna(median_versions_per_item_property) and median_versions_per_item_property > 1:
    implications.append(
        "Свойства товаров обновляются во времени, значит при построении признаков нужно делать temporal snapshot и не допускать утечки из будущего"
    )
else:
    implications.append(
        "Свойства товаров выглядят относительно стабильными и подходят для item feature engineering"
    )

if transaction_after_cart_rate > 20:
    implications.append(
        f"Пары user-item с add-to-cart имеют сильную связь с покупкой: conversion cart->transaction = {transaction_after_cart_rate}%. "
        f"Это подтверждает бизнес-смысл оптимизации add-to-cart"
    )
else:
    implications.append(
        f"Конверсия cart->transaction = {transaction_after_cart_rate}%, поэтому стоит аккуратно проверить согласованность событий и таймлайн"
    )

implications.append(
    "Для офлайн-оценки лучше делать time-based split, а не случайное перемешивание, так как данные событийные и зависят от времени"
)

implications.append(
    "Минимально разумный стек сравнения моделей: popularity baseline, item-to-item co-occurrence или implicit collaborative filtering, затем более сильный ranker для reranking кандидатов"
)

event_distribution_lines = []
for event_name, cnt in event_counts.items():
    share = event_shares.loc[event_name]
    event_distribution_lines.append(f"- {event_name}: {format_num(cnt)} строк, {share}%")

markdown = f"""# EDA Findings

## 1. Data overview

### Tables
- `category_tree.csv`: {format_num(category_rows)} строк, {category_tree.shape[1]} столбцов
- `events.csv`: {format_num(events_rows)} строк, {events_cols} столбцов
- `item_properties.csv` после объединения частей: {format_num(item_props_rows)} строк, {item_props_cols} столбцов

### Missing values
- `category_tree`: {format_num(category_missing)} пропусков всего
- `events`: {format_num(events_missing)} пропусков всего
- `item_properties`: {format_num(item_props_missing)} пропусков всего

### Uniques
- уникальных пользователей в `events`: {format_num(unique_users)}
- уникальных товаров в `events`: {format_num(unique_items_events)}
- уникальных transaction id: {format_num(unique_transactions)}
- уникальных товаров в `item_properties`: {format_num(unique_items_props)}
- уникальных свойств товаров: {format_num(unique_properties)}
- уникальных значений свойств: {format_num(unique_property_values)}

### Category tree
- уникальных parent category: {format_num(category_unique_parents)}
- уникальных child category: {format_num(category_unique_children)}

## 2. Time coverage

- период `events`: {format_dt(events_min_ts)} -> {format_dt(events_max_ts)}
- период `item_properties`: {format_dt(props_min_ts)} -> {format_dt(props_max_ts)}

## 3. Event structure

### Event distribution
{chr(10).join(event_distribution_lines)}

### Daily summary
{daily_events_md}

## 4. User behavior

- медианное число событий на пользователя: {format_num(median_events_per_user)}
- 95-й перцентиль событий на пользователя: {format_num(p95_events_per_user)}
- 99-й перцентиль событий на пользователя: {format_num(p99_events_per_user)}

Интерпретация: если 95-й и 99-й перцентили сильно выше медианы, в данных выражен long tail по активности пользователей.

## 5. Item behavior

- медианное число событий на товар: {format_num(median_events_per_item)}
- 95-й перцентиль событий на товар: {format_num(p95_events_per_item)}
- 99-й перцентиль событий на товар: {format_num(p99_events_per_item)}
- доля событий, приходящаяся на топ-10 товаров: {top_10_items_share}%
- доля событий, приходящаяся на топ-100 товаров: {top_100_items_share}%

Интерпретация: высокая концентрация на небольшом числе товаров означает, что popularity baseline будет сильным и его обязательно нужно включить в сравнение.

## 6. Conversion patterns

- всего уникальных пар user-item: {format_num(pairs_total)}
- пар с просмотром: {format_num(pairs_with_view)}
- пар с add-to-cart: {format_num(pairs_with_cart)}
- пар с transaction: {format_num(pairs_with_transaction)}
- доля пар с add-to-cart среди просмотренных пар: {cart_after_view_rate}%
- доля пар с transaction среди просмотренных пар: {transaction_after_view_rate}%
- доля пар с transaction среди пар с add-to-cart: {transaction_after_cart_rate}%
- доля пар user-item с повторными просмотрами: {repeat_views_share}%
- доля пар user-item с повторными добавлениями в корзину: {repeat_carts_share}%

## 7. Item properties

- медианное число версий на пару item-property: {format_num(median_versions_per_item_property)}
- 95-й перцентиль числа версий на пару item-property: {format_num(p95_versions_per_item_property)}
- товаров только в `events`, но не в `item_properties`: {format_num(items_only_in_events)}
- товаров только в `item_properties`, но не в `events`: {format_num(items_only_in_props)}

### Top properties
{top_properties_md}

Интерпретация: множественные версии свойств по времени означают, что для обучения и инференса нужен снимок признаков на момент события, а не последняя версия из будущего.

## 8. Implications for recommender design

{chr(10).join(f"- {x}" for x in implications)}

## 9. Recommended offline setup

- Целевой сигнал: `addtocart` как основной позитивный event
- Дополнительный сильный позитивный сигнал: `transaction`
- Исключение из рекомендаций: уже купленные товары пользователя
- Разбиение: строго по времени, train на прошлом, validation на будущем
- Базовые метрики: Recall@K, HitRate@K, MAP@K или NDCG@K
- Бизнес-логика: метрики должны отражать способность модели возвращать товары, которые пользователь добавит в корзину
- Минимальный набор baseline-моделей:
  - popularity
  - co-occurrence item-to-item
  - collaborative filtering или retrieval-кандидаты
  - ranker для reranking кандидатов, если успевает по объему проекта

## 10. Risks and constraints

- Возможна сильная разреженность user-item матрицы
- Возможен cold start по товарам и пользователям
- Свойства товаров могут быть шумными и обновляться во времени
- При случайном split возникнет временная утечка
- Нужно заранее определить, какие события считаются позитивными, а какие использовать только как контекст
"""

OUTPUT_PATH.write_text(markdown, encoding="utf-8")

print(f"Готово. Файл сохранён в: {OUTPUT_PATH.resolve()}")
print()
print(markdown[:4000])
```
